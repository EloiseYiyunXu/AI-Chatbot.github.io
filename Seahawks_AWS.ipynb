```python
import pandas as pd
import chardet

# Detect encoding
with open('seahawks_draft_links_Final_1.csv', 'rb') as file:
    result = chardet.detect(file.read())  # or read a portion of the file file.read(10000)
encoding = result['encoding']

# Read file with detected encoding
df = pd.read_csv('seahawks_draft_links_Final_1.csv', encoding=encoding)
```

```python
#Don't run this code!
metadata = []
content = []

for index, row in df.iterrows():
    content.append(str(row['Content']))
    metadata.append({
        'title': row['Title'],
        'URL':row['URL']
    })

text_splitter = CharacterTextSplitter(chunk_size=800, chunk_overlap=0)
docs = text_splitter.create_documents(content, metadata)
```

```python
#Don't run this code!
documents = vector_store.from_documents(
    docs,
    es_cloud_id=CLOUD_ID,
    es_user=CLOUD_USERNAME,
    es_password=CLOUD_PASSWORD,
    index_name="seahawk6",
    strategy=ElasticsearchStore.SparseVectorRetrievalStrategy()
)
```

```python
default_model_id = "amazon.titan-text-express-v1"
AWS_MODEL_ID = input(f"AWS model [default: {'amazon.titan-text-express-v1'}]: ") or default_model_id
llm = Bedrock(
    client=bedrock_client,
    model_id= AWS_MODEL_ID
)


```

```python
retriever = vector_store.as_retriever()

qa = RetrievalQA.from_llm(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)

questions = [
    'Should the Seahawks Draft Michael Penix, why and with what pick?'
]
question = questions[0]
print(f"Question: {question}\n")

ans = qa({"query": question})

print("\033[92m ---- Answer ---- \033[0m")
print(ans["result"] + "\n")
print("\033[94m ---- Sources ---- \033[0m")
for doc in ans["source_documents"]:
  print("Name: " + doc.metadata["title"])
  print("Content: "+ doc.page_content)
  print("-------\n")
```
